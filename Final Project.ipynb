```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
cuba_1 = pd.read_excel("LB_Cuba(1)09_12_1.XLSX")
cuba_1.rename(columns={'Normalized Pk Intensities': 'normalized_pk_intensities'}, inplace=True)
cuba_1= cuba_1.set_index("normalized_pk_intensities")
cuba_1.head()
```

```python
columns_to_drop = [9]+ [10] +list(range(5, 8)) 
cuba_1 = cuba_1.drop(cuba_1.columns[columns_to_drop], axis=1) # here I drop the MgO%, TiO%, SiO%, FeO% and MnO% columns
cuba_1.head()
```

```python
column_means_1 = cuba_1.mean() # Taking the mean of all elements values (and their isotopes eg. Li.1, B.1, Nd.1)
column_means_1
```

```python
# Ploting my graph
plt.figure(figsize=(12, 6))
column_means_1.plot(kind="bar", color="skyblue", edgecolor="black")

# labels and title
plt.title("Mean Values of Elements in Cuba 1 Sample")
plt.xlabel("Chemical Elements")
plt.ylabel("Average concentration in ppm")
plt.xticks(rotation=0)  # Keeping the x-axis labels horizontal
plt.grid(axis="y", linestyle="--", alpha=0.7)  # horizontal gridlines
plt.show()
```

#### Explanation of my graph:
#### Cr, Ni and Co are typically associated with ultramafic rocks derived from the mantle, such as peridotites.
#### High Cr concentrations in serpentinite suggest that the parent mineral could be pyroxenes.
#### This indicates that the serpentinite likely originated from mantle material rather than oceanic crust.
#### During partial melting, Ni and Co are preferentially retained in the residue, which may result in higher concentrations in ultramafic rocks.
#### Sr (Strontium) is less common in mantle rocks but can be introduced during interaction with crustal materials or fluids.

```python
# Comparing my RRE concentrations to a reference standard- Ploting REE using normalization (Chondritic Meteorite- M&S 1995): 
# Givens (I took the data from https://meteorites.wustl.edu/goodstuff/ree-chon.htm)
chondrite_values={
    "Y": 1.57,
    "La": 0.237, 
    "Ce": 0.613, 
    "Pr": 0.0928, 
    "Nd": 0.457,
    "Sm": 0.148,
    "Eu": 0.0563,
    "Gd": 0.199,
    "Tb": 0.0361,
    "Dy": 0.246,
    "Ho": 0.0546,
    "Er": 0.160,
    "Tm": 0.0247,
    "Yb": 0.161,
    "Lu": 0.0246}

matching_keys_REE_1 = [key for key in chondrite_values.keys() if key in cuba_1.columns]

column_means_REE_1 = cuba_1[matching_keys_REE_1].mean()
ratios_1 = {key: chondrite_values[key] / column_means_REE_1[key] for key in matching_keys_REE_1}

# Ploting the results
plt.figure(figsize=(10, 5))
plt.plot(list(ratios_1.keys()), list(ratios_1.values()), marker="o", linestyle="-", color="blue", label="Ratio")

# Add labels, title, and grid
plt.title("Normalized REE")
plt.xlabel("REE")
plt.ylabel("Normalized Ratio")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.show()
```

#### Explanation:
#### Relative high levels of Nd, Gd, and Dy suggest the source rock may have been influenced by processes that enriched REEs, 
#### such as partial melting or mantle metasomatism
#### In addition, subducted sediments or crust can release REEs into fluids that infiltrate the mantle wedge
